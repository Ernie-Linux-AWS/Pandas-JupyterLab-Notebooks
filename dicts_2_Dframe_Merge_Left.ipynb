```python
'''
'dicts_2_Dframe_Merge_Left.ipynb'      

Create dictionaries 'dic_a' and 'dic_b'.

Create DataFrame 'df_a' from dictionary 'dic_a'.
Create DataFrame 'df_b' from dictionary 'dic_b'.

Left Merging:

Create a DataFrame via a merge of df_a and df_b
    with common values in the same rows in
    column 'Position'.

Create a DataFrame via a merge of df_a and df_b
    with common values in the same rows in
    columns 'Position' and 'Value'.

Create a DataFrame via a merge of df_a and df_b
    with common values in the same rows in
    columns 'Position', 'Value' and 'Extra'.

Delete all created DataFrames from memory.
'''
#
from IPython.display import display
from pathlib import Path
import os
import pandas as pd
import time
#

# Create dictionaries 'dic_a' and 'dic_b'.
# Create DataFrame 'df_a' from dictionary 'dic_a'.
# Create DataFrame 'df_b' from dictionary 'dic_b'.
dic_a = {
    'Position': [3, 4, 5, 6],
    'Value': ['a', 'x', 'y', 'a'],
    'Extra': ['e', 'o', 'j', 'r',],
}
dic_b = {
    'Position': [3, 4, 7, 8],
    'Value': ['a', 'x', 'z', 'b'],
    'Extra': ['e', 'o', 'k', 's',],
}
df_a = pd.DataFrame(dic_a)
df_b = pd.DataFrame(dic_b)
print('df_a.info():')
display(df_a.info())
display(df_a)
print('\ndf_b.info():')
display(df_b.info())
display(df_b)

# Return a DataFrame with common values in the same rows in
#   column 'Position'.
on_1 = 'Position'

# Return a DataFrame with common values in the same rows in
#   columns 'Position' and 'Value'
on_2 = ['Position', 'Value']

# Return a DataFrame with common values in the same rows in
#   columns 'Position' and 'Value'
on_3 = ['Position', 'Value', 'Extra']

# Merge types:
# merge_how = 'inner'
merge_how = 'left'
# merge_how = 'right'
# merge_how = 'outer'

# Create a DataFrame via a merge of df_a and df_b
#   with common values in the same rows in
#   columns 'Position'.
print(f"\nMerge Type: '{merge_how}'\n")
df_inner1 = df_a.merge(df_b, on=on_1, how=merge_how)
display(df_inner1)
if df_inner1.empty:
    print('Empty DataFrame found')
    df_inner1.info()

# Create a DataFrame via a merge of df_a and df_b
#   with common values in the same rows in
#   columns 'Position' and 'Value'.
df_inner2 = df_a.merge(    df_b, on=on_2, how=merge_how)
display(df_inner2)
if df_inner2.empty:
    print('Empty DataFrame found')
    df_inner2.info()

# Create a DataFrame via a merge of df_a and df_b
#   with common values in the same rows in
#   columns 'Position', 'Value' and 'Extra'.
df_inner3 = df_a.merge(    df_b, on=on_3, how=merge_how)
display(df_inner3)
if df_inner3.empty:
    print('Empty DataFrame found')
    df_inner3.info()

# Delete DataFrames 'df_a', 'df_b', df_inner1, df_inner2 and
#   df_inner3 from memory.
del df_a
del df_b
del df_inner1
del df_inner2
del df_inner3

# Get time info.
time_local = time.localtime()
time_string = time.strftime('%Y-%m-%d %H:%M:%S %Z %z', time_local)
print(f'\n{time_string}')
# EOF
```
