```python
'''
'dict_2_Dframe_2_xlsx.ipynb'      

Create a DataFrame from a dictionary.  
Rename the Index and all Index Labels.   

Add a labeled row at a precise location.
Rename the Index Label.

Concatenate new rows with specific Index labels.
Drop labeled rows.

Add labeled columns at a precise locations.  
Drop a column.

Write DataFrame 'df_carsales' as a CSV file.
Write DataFrame 'df_carsales' as an xlsx file.

Delete DataFrame 'df_carsales' from memory.
'''
#
from IPython.display import display
from pathlib import Path
import os
import pandas as pd
import time
#

# Create dictionary'dic_cardata'.
# Display DataFrame 'df_carsales' data types.
# Create DataFrame 'df_carsales' from dictionary 'dic_cardata'.
# Rename the Index of DataFrame 'df_carsales'.
# Print a row via an index
# Display DataFrame 'df_carsales'.
dic_cardata = {
    'Mercedes': [2, 4, 0, 4, 0, 3],
    'Ford': [3, 0, 0, 1, 6, 12],
    'BYD': [9, 3, 4, 1, 0, 0],
    'Renault': [12, 1, 0, 0, 3, 1],
}
df_carsales = pd.DataFrame(dic_cardata)
print(df_carsales.info())
df_carsales.index.rename(
    'Sales Location', inplace=True
)

# Rename the Index Labels of DataFrame 'df_carsales'.
# Print a row via a renamed Index Label.
# Display DataFrame 'df_carsales'.
print(f'df_carsales.loc[1]\n{df_carsales.loc[1]}')
display(df_carsales)
new_idx_dic = {
    0: 'One', 1: 'Two', 2: 'Three', 3: 'Four', 4: 'Five', 5:
        'Six',
}
df_carsales.rename(
    index=new_idx_dic, inplace=True
)

print(f"df_carsales.loc['Two']\n{df_carsales.loc['Two']}")
display(df_carsales)

# Add a labeled row at a precise location.
# Rename the Index Label.
new_row_idx = 7
new_row_idx_name = 'Seven'
new_row = {
    'Mercedes': 3, 'Ford': 5, 'BYD': 1, 'Renault': 1,
}
df_carsales.loc[new_row_idx] = new_row
df_carsales = df_carsales.rename(
    index={
        new_row_idx: 'Seven',
    }
)
print(f"df_carsales.loc['Seven']\n{df_carsales.loc['Seven']}")
display(df_carsales)

# Concatenate a new row with a specific Index Label.
new_row_dict = {
    'Mercedes': 8, 'Ford': 8, 'BYD': 8, 'Renault': 8,}
new_row = pd.DataFrame(
    new_row_dict,  index=['Eight']
)
                       
df_carsales = pd.concat([df_carsales, new_row])
print(f"df_carsales.loc['Eight']\n{df_carsales.loc['Eight']}")
display(df_carsales)

# Concatenate a new row with a specifc Index Label.
new_row_dict = {
    'Mercedes': 9, 'Ford': 9, 'BYD': 9, 'Renault': 9,}
new_row = pd.DataFrame(
    new_row_dict,  index=['Nine']
)

df_carsales = pd.concat([df_carsales, new_row])
print(f"df_carsales.loc['Nine']\n{df_carsales.loc['Nine']}")
display(df_carsales)

# Drop labeled row.
row_to_drop = 'Nine'
df_carsales.drop(
    index=row_to_drop, axis='index', inplace=True)
display(df_carsales)

# Drop labeled row.
row_to_drop = 'Eight'
df_carsales.drop(
    index=row_to_drop, axis='index', inplace=True)
display(df_carsales)

# Add a trailing labeled column.
# Insert new labeled columns at  precise locations.
# Drop column 'Toyota'.
Hyundai = [2, 6, 2, 3, 0, 4, 7]
Mazda = [12, 11, 0, 0, 0, 4,1]
Toyota = [2, 3, 4, 3, 2, 0, 1]
Volvo = [1, 4, 0, 0, 0, 11,12]
df_carsales['Volvo'] = Volvo
display(df_carsales)
df_carsales.insert(loc=0, column='Hyundai', value=Hyundai)
display(df_carsales)
df_carsales.insert(loc=1, column='Mazda', value=Mazda)
display(df_carsales)
df_carsales.insert(loc=2, column='Toyota', value=Mazda)
display(df_carsales)
columns_to_drop = ['Toyota']
df_carsales.drop(
    columns=columns_to_drop, axis='column', inplace=True
)
display(df_carsales)

# Ensure that the target directory exists.
# Ensure that the target file is writeable.
# Write DataFrame 'df_carsales' as a CSV file while
#   including the Index and Index Labels (default).
# Update './data' datetime stamp
name_dir = './data'
name_filepath = './data/carsales.csv'
os.makedirs(name_dir, exist_ok=True)
if os.path.exists(name_dir):
    os.chmod(name_dir, 0o755)    
if os.path.exists(name_filepath):
    os.chmod(name_filepath, 0o644)
with open(
    name_filepath, 'w'
) as writer:
    df_carsales.to_csv(writer, index_label='Sales Location')
    if os.path.exists(name_filepath):
        os.chmod(name_filepath, 0o644)
        print(df_carsales.info())

        # Get then display modified time info
        ti_m = os.path.getmtime(name_filepath)
        m_ti = time.ctime(ti_m)
        Path(name_dir).touch()
        print(f"File:\n'{name_filepath}' was created or modified: {m_ti}")

# xlsx read/write initialization requirements:
#   1. pip install 'openpyxl'.
#   2. Then a kernel restart MAY be required ONCE.
#
# Ensure that the target directory exists.
# Ensure that the target file is writeable.
# Write DataFrame 'df_carsales' as an xlsx file while
#   including the Index and Index Labels (default).
# Update './data' datetime stamp
name_dir = './data'
name_filepath = './data/carsales.xlsx'
os.makedirs(name_dir, exist_ok=True)
if os.path.exists(name_dir):
    os.chmod(name_dir, 0o755)
if os.path.exists(name_filepath):
    os.chmod(name_filepath, 0o644)
with pd.ExcelWriter(
    name_filepath
) as writer:
    df_carsales.to_excel(writer, index_label='Sales Location')
    if os.path.exists(name_filepath):
        os.chmod(name_filepath, 0o444)
        
        # Get then display modified time info
        ti_m = os.path.getmtime(name_filepath)
        m_ti = time.ctime(ti_m)
        Path(name_dir).touch()
        print(f"File:\n'{name_filepath}' was created or modified: {m_ti}")

# Delete DataFrame 'df_carsales' from memory.
del df_carsales

# Get time info.
time_local = time.localtime()
time_string = time.strftime('%Y-%m-%d %H:%M:%S %Z %z', time_local)
print(f'\n{time_string}')
# EOF
```
