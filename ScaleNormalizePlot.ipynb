```python
'''
'ScaleNormalizePlot.ipynb'  

Create a DataFrame by loading seaborn dataset 'geyser'.

Create a pair of columns where:
    data is means corrected.

Create a pair of columns where:
    data is normalized using a standard deviation of one.

Create a pair of columns where:
   data values are rescaled between 0 - 1.

Plot:
   Original  DataFrame pair,
   Means corrected DataFrame pair,
   Normalized DataFrame pair, and the
   Zero-One DataFrame pair
   as a single column of subplots.

Plot:
   Original  DataFrame pair,
   Means corrected DataFrame pair,
   Normalized DataFrame pair, and the
   Zero-One DataFrame pair
   as a double row of subplots where each row contains two plots.
'''
#
from IPython.display import display
from pathlib import Path
import matplotlib.pyplot as plt
import os
import seaborn as sns
import pandas as pd
import time
#
geyser = sns.load_dataset('geyser')
display(geyser)

# Create a pair of columns where:
#   data is means corrected.
geyser['mnc_duration'] = geyser['duration'] - geyser['duration'].mean()
geyser['mnc_waiting'] = geyser['waiting'] - geyser['waiting'].mean()
display(geyser)

# Create a pair of columns where:
#   data is normalized using a standard deviation of one.
geyser['std_duration'] = (
    geyser['duration'] - geyser['duration'].mean()
    ) / geyser['duration'].std()
geyser['std_waiting'] = (
    geyser['waiting'] - geyser['waiting'].mean()
    ) / geyser['waiting'].std()
display(geyser)

# Create a pair of columns where
#   data values are rescaled between 0 - 1.
geyser['Z1_duration'] = (
    geyser['duration'] - geyser['duration'].min()
    ) / ( geyser['duration'].max() - geyser['duration'].min()
)
geyser['Z1_waiting'] = (
    geyser['waiting'] - geyser['waiting'].min()
    ) / (geyser['waiting'].max() - geyser['waiting'].min()
)
display(geyser)

# Plot:
#   Original  DataFrame pair,
#   Means corrected DataFrame pair,
#   Normalized DataFrame pair, and the
#   Zero-One DataFrame pair
#   as a single column of subplots.
hspace = 0.3
ncols =1
nrows =4
wspace = 0.4
height_size = 15
width_size = 3.2
# (Each value in the 'axes' represents a comma separated row).
fig, (orig, mc, std, Z1) = plt.subplots(
    nrows=nrows, ncols=ncols, figsize=(width_size,height_size)
)
geyser.plot('duration', 'waiting', kind='scatter', ax=orig)
geyser.plot('mnc_duration', 'mnc_waiting', kind='scatter', ax=mc)
geyser.plot('std_duration', 'std_waiting', kind='scatter', ax=std)
geyser.plot('Z1_duration', 'Z1_waiting', kind='scatter', ax=Z1)
# Adjust spacing between plots.
plt.subplots_adjust(hspace=hspace)
plt.show()

# Plot:
#   Original  DataFrame pair,
#   Means corrected DataFrame pair,
#   Normalized DataFrame pair, and the
#   Zero-One DataFrame pair
#   as a double row of subplots where each row contains two plots.
hspace = 0.5
ncols =2
nrows =2
height_size = 8
width_size = 8
# (Each tuple in the 'axes' represents a comma separated row).
fig, ((orig, mc), (std, Z1)) = plt.subplots(
    nrows=nrows, ncols=ncols, figsize=(width_size,height_size)
)
geyser.plot('duration', 'waiting', kind='scatter', ax=orig)
geyser.plot('mnc_duration', 'mnc_waiting', kind='scatter', ax=mc)
geyser.plot('std_duration', 'std_waiting', kind='scatter', ax=std)
geyser.plot('Z1_duration', 'Z1_waiting', kind='scatter', ax=Z1)
# Adjust spacing between plots.
plt.subplots_adjust(hspace=hspace, wspace=wspace)
plt.show()

# Get time info.
time_local = time.localtime()
time_string = time.strftime('%Y-%m-%d %H:%M:%S %Z %z', time_local)
print(f'\n{time_string}')
# EOF
```
